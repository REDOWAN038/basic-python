```python
import pandas as pd
```

```python
bd_weather = pd.DataFrame({
 "city":["dhaka", "sylhet", "ctg"],
 "temperature":[30,32,34],
 "humidity":[68,65,70]
})
bd_weather
```

```python
usa_weather = pd.DataFrame({
 "city":["new york", "chicago", "LA"],
 "temperature":[12,14,15],
 "humidity":[21,14,35]
})
usa_weather
```

```python
df = pd.concat([bd_weather, usa_weather])
df
```

```python
df = pd.concat([bd_weather, usa_weather], ignore_index=True)
df
```

```python
df = pd.concat([bd_weather, usa_weather], keys=["bd", "us"])
df
```

```python
df.loc["bd"]
```

```python
temp_weather = pd.DataFrame({
 "city":["dhaka", "sylhet", "ctg"],
 "temperature":[30,32,34],
})
temp_weather
```

```python
hum_weather = pd.DataFrame({
 "city":["dhaka", "sylhet", "ctg"],
 "humidity":[68,65,70]
})
hum_weather
```

```python
df = pd.concat([temp_weather, hum_weather], axis=1)
df
```
