```python
import matplotlib.pyplot as plt

class Circle(object):
  def __init__(self, radius=3, colour='red'):
    self.radius = radius
    self.colour = colour

  def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.colour))
        plt.axis('scaled')
        plt.show() 

C1 = Circle(5,'blue')
C1.drawCircle()
C2 = Circle()
C2.drawCircle()
```
