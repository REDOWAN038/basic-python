```python
import pandas as pd

songs = {'Album' : ['Bat Out of Hell','Thriller','Back in Black','The Bodyguard'],
         'Released' : [1977,1982,1980,1992],
         'Length' : ['46:33','42:19','42:11','57:44']}

# creating a dataframe
songsFrame = pd.DataFrame(songs)
songsFrame.head()
```

```python
# creating data frame from songsFrame

albumFrame = songsFrame[['Album']]
albumFrame.head()
```

```python
# songs released after 1979

df = songsFrame[songsFrame['Released']>=1980]
df.head()
```

```python
# saving as csv file
df.to_csv('new_songs.csv')
```

```python
# how to read data from csv file

csv_path = 'new_songs.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# iloc() ---> you have to give index to access value
df.iloc[0,1]
```

```python
# loc() ---> you have to give row or coloumn name
# first parameter must be index name
df.loc[0,'Released']
```

```python
# return values in coloumn

df2 = df.loc[:, ['Album','Length']]
df2.head()
```

```python
df3 = df.iloc[0:2,0:2]
df3.head()
```
